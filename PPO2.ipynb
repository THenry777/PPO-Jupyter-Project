## Univariate analysis (UA)
UA is a kind of analysis where we look at the characteristics of a single column's data. In our case, since the columns are all  contain continuous numerical values, we use histograms to examine the distribution.

### Step 3: Perform UA on 'AT' using histogram
First let's examine ambient temperature using a histogram. 

```python
# Step 3: Plot histogram with 'AT'
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_excel('Folds5x2_pp.ods')
df.hist(column='AT', grid=False, edgecolor='black')
print(df)
```

### Step 8: Plot a pairplot
Since we have a small number of columns, and these columns contain only continuous numerical values, we can plot a pairplot. A pairplot is a <em>m</em> x <em>m</em> grid containing scatterplots, where <em>m</em> is the number of columns in the DataFrame. 

The diagonal from top left to lower right contains the column's histogram, whereas the rest are scatterplots with the combinations of columns.



Plot a pairplot with the DataFrame column values.

After you plot, pay attention to any possible linear looking patterns at the last row, i.e PE vs other columns.

<strong>Hint: use seaborn</strong>

```python
# Step 8: Plot a pairplot
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_excel('Folds5x2_pp.ods')
sns.pairplot(df)
plt.show()
```
